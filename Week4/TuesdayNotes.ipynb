__Electric (Coulumb) Potential of a charge dipole.__

\begin{equation}
V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}} - \frac{kq}{\sqrt{x^2+(y+d)^2}}
\end{equation}

where $k = \frac{1}{4 \pi \epsilon_o} $

```python
# make array with x coordinates spaced along.d
import numpy as np
dx = 0.06
dy = 0.06
x,y = np.meshgrid(np.arange(-2.,2.,dx),np.arange(-2.,2.,dy))

print x.ndim
print x.shape
print y.ndim
print y.shape
print x is y

print x[:5,:10]
print y[:5,:10]

A = np.ones((10,5),dtype='int')
print A.shape
print len(A[0])
print len(A[0,:])
# print len(A[0,0]) # Last multi-slice argument is the inner most array elements.
```

The `np.meshgrid` produces a tuple of arrays. Each array has an element for every point in the space, but only one of the coordinates of that point.

```python
# make the function V(x,y)
k=1
q=1
d=1

#def calculate_V(x,y,k,q,d):
#    """x, y are equally shaped ndarrays. returns a new array with the same shape each elelment the 
#    result of evaluating the dipole potential for the respective x and y elements.
#    Also needs k,q,d specified."""
#    if type(x) != ndarray or type(y) != ndarray:
#        raise 'Argmuent type invalid'
#        return None
#    elif x.shape != y.shape:
#        return None
#    V = np.empty(x.shape,dtype='float')
#    for i in range(len(V)):
#        for j in range(len(V[0])):
#            V[i,j] = (k*q)/(x[i,j]**2. + (y[i,j]-d)**2.)**0.5 - (k*q)/(x[i,j]**2. + (y[i,j]+d)**2.)**0.5
#    return V

V = (k*q)/(x**2. + (y-d)**2.)**0.5 - (k*q)/(x**2. + (y+d)**2.)**0.5
```
