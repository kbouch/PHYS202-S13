__Electric (Coulumb) Potential of a charge dipole.__

\begin{equation}
V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}} - \frac{kq}{\sqrt{x^2+(y+d)^2}}
\end{equation}

where $k = \frac{1}{4 \pi \epsilon_o} $
