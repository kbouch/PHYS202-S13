#Example 1

```python
from scipy.optimize import curve_fit

curve_fit?
```

```python
def func(x,a,b,c):
    return a*x**2 + b*x + c

x = np.array([0.,1.,2.,3.,4.,5.])
y = np.array([-1.78,4.09,8.85,17.9,26.1,35.2])
yerr = np.array([0.460,0.658,0.528,1.34,1.09,0.786])
```

```python
p0 = [-1.,3.,-2.]
```

```python
popt,pcov = curve_fit(func,x,y,p0,yerr)

p_unc = np.zeros(len(popt),dtype=float)
for i in range(len(popt)):
    p_unc[i] = np.sqrt(pcov[i,i])
```

```python
print 'optimal parameter: ',popt
print 'uncertainties in parameters: ',p_unc
```

```python
fig = gcf()
fig = figure(figsize=(8.,8.))
ax = gca()
ax.errorbar(x,y,yerr,fmt='r*')

x_dom = np.arange(0.,5.,0.01)
fitresult = func(x_dom,popt[0],popt[1],popt[2])
ax.plot(x_dom,fitresult,'b-')
```

#Example 2

```python
waveform_1 = np.load('waveform_1.npy')
t = np.arange(len(waveform_1))

plot(t,waveform_1)
xlabel('Time (ns)',fontsize = 20.)
ylabel('Intensity (bins)',fontsize = 20.)
```

Fit with a gaussian against a background level.

Form $ B + A \operatorname{exp} \Bigg[ - \Bigg({\displaystyle \frac{t- \mu}{\sigma} }\Bigg ) ^{2} \Bigg] $

```python
def model_1(t,B,A,mu,sigma):
    return B + (A*np.exp(- ((t-mu)/sigma)**2))
```

```python
p0 = np.array([4.,25.,15.,5.])
```

```python
p_opt,p_cov = curve_fit(model_1,t,waveform_1,p0)

plot(t,waveform_1,label='data')
plot(t,model_1(t,p_opt[0],p_opt[1],p_opt[2],p_opt[3]),label='fit',\
     linewidth=2.)
legend(loc='best')
xlabel('Time (ns)',fontsize = 20.)
ylabel('Intensity (bins)',fontsize = 20.)
```
