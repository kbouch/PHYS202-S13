###Example 1

Integrate $ f(x) = x ^{1/2} + \sin (x) \quad $.

```python
import numpy as np
f = lambda x: x**0.5 + np.sin(x)
```

[Numpy element-wise Universal Functions list][1]

   [1]:http://docs.scipy.org/doc/numpy/reference/ufuncs.html#available-ufuncs

```python
xmin = 0.
xmax = 15.

x = np.arange(xmin,xmax,0.001)

ax = gca()
my_plot = ax.plot(x,f(x),'b-',linewidth=3.,label=r'$f(x)$')
xlabel(r'$x$',fontsize=20.)
ylabel(r'$ f(x) = x ^{1/2} + \sin (x)$',fontsize=20.)

xlim(0.,15.)

show()
```

Integrate between 1.7,12.4

```python
xmin2 = 1.7
xmax2 = 12.4
subx = x[logical_and(x>xmin2,x<xmax2)]
ymin2 = 0.
ymax2 = 1.05*max(f(subx))

ax = gca()
my_plot = ax.plot(x,f(x),'b-',linewidth=3.,label=r'$f(x)$')
xlim(0.,15.)


ax.plot( [xmin2,xmin2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmax2,xmax2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmin2,xmax2] , [ymax2,ymax2] , color='k',linestyle='--')

show()
```

```python
N = 10000
samples_x = xmin2 + (xmax2 - xmin2)*np.random.random_sample(N)
samples_y = ymin2 + (ymax2 - ymin2)*np.random.random_sample(N)

ax = gca()
my_plot = ax.plot(x,f(x),'b-',linewidth=3.,label=r'$f(x)$')
xlim(0.,15.)

ax.plot(samples_x[:200],samples_y[:200],'g.')

ax.plot( [xmin2,xmin2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmax2,xmax2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmin2,xmax2] , [ymax2,ymax2] , color='k',linestyle='--')

show()
```

```python
MC_mask = (samples_y < f(samples_x))

print MC_mask.shape


n_in = 0
for elem in MC_mask:
    if elem == True:
        n_in += 1

print n_in
n_out = MC_mask.shape[0] - n_in
print n_out
```

```python
ax = gca()
my_plot = ax.plot(x,f(x),'b-',linewidth=3.,label=r'$f(x)$')
xlim(0.,15.)

ax.plot(samples_x[:200],samples_y[:200],'g.')
ax.plot(samples_x[MC_mask[:200]],samples_y[MC_mask[:200]],'r.')

ax.plot( [xmin2,xmin2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmax2,xmax2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmin2,xmax2] , [ymax2,ymax2] , color='k',linestyle='--')

show()
```

```python
I_approx = (n_in/float(MC_mask.shape[0]))*((xmax2 - xmin2)*(ymax2 - ymin2))
print 'Monte Carlo approximation: ',I_approx

fprime = lambda x: (2/3.0)*x**(3/2.0) - np.cos(x)

I_exact = fprime(12.4) - fprime(1.7)
print '\nExact value: ',I_exact
print '\npercent error = ',(I_exact - I_approx)*100.0/I_exact,'%'
```

```python
# N =    100,000  %error = 0.115
# N =  1,000,000  %error = 0.082085
# N =  5,000,000  %error = 0.023376
# N = 10,000,000  %error = 0.003559
# N = 12,000,000  %error =-0.00703  # worse?  different random sample

x = np.arange(0.,15.,0.001)

xmin2 = 1.7
xmax2 = 12.4
subx = x[logical_and(x>xmin2,x<xmax2)]
ymin2 = 0.
ymax2 = 1.05*max(f(subx))

N = 12000000
samples_x = xmin2 + (xmax2 - xmin2)*np.random.random_sample(N)
samples_y = ymin2 + (ymax2 - ymin2)*np.random.random_sample(N)



ax = gca()

ax.plot(x,f(x),'b-',linewidth=3.,label=r'$f(x)$')
ax.plot(samples_x[:200],samples_y[:200],'g.')

MC_mask = (samples_y < f(samples_x))

n_in = 0
for elem in MC_mask:
    if elem == True:
        n_in += 1

ax.plot(samples_x[MC_mask[:200]],samples_y[MC_mask[:200]],'r.')

xlim(0.,15.)


ax.plot( [xmin2,xmin2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmax2,xmax2] , [ymin2,ymax2] , color='k',linestyle='--')
ax.plot( [xmin2,xmax2] , [ymax2,ymax2] , color='k',linestyle='--')

show()


I_approx = (n_in/float(MC_mask.shape[0]))*((xmax2 - xmin2)*(ymax2 - ymin2))
print 'Monte Carlo approximation: ',I_approx

fprime = lambda x: (2/3.0)*x**(3/2.0) - np.cos(x)

I_exact = fprime(12.4) - fprime(1.7)
print '\nExact value: ',I_exact
print '\npercent error = ',(I_exact - I_approx)*100.0/I_exact,'%'
```

```python
import numpy as np
f = lambda x: x**0.5 + np.sin(x)
    
fprime = lambda x: (2/3.0)*x**(3/2.0) - np.cos(x)
    
```

```python
# N =     10,000  %error = 
# N =    100,000  %error = 0.115
# N =  1,000,000  %error = 0.082085
# N =  5,000,000  %error = 0.023376
# N = 10,000,000  %error = 0.003559
# N = 12,000,000  %error =-0.00703

x = np.arange(0.0,15.0,0.001)
subx = x[logical_and(x>1.7,x<12.4)]

xmin = 1.7
xmax = 12.4

def mean_at_sampled(sample_arr,func):
    Sum = func(sample_arr).sum()
    N = float(sample_arr.size)
    return Sum/N


N = 100000
samples_x = xmin2 + (xmax - xmin)*np.random.random_sample(N)

mean_value = mean_at_sampled(samples_x,f)

I_approx = (xmax-xmin)*mean_value

print 'Sample mean method approximation:',I_approx
I_exact = fprime(12.4) - fprime(1.7)
print '\nExact value: ',I_exact
print '\npercent error = ',(I_exact - I_approx)*100.0/I_exact,'%'

AA = np.empty(4,dtype='int')
AA.fill(12)
print AA
print AA[:2]

S = subx.shape

#####mean_value_arr = np.empty(10,dtype='float').fill(mean_value)   ????????   .fill() not working?

#####print mean_value_arr   ??????   Gives me None

ax = gca()

ax.plot(x,f(x),'b-',linewidth=3.,label=r'$f(x)$')
ax.plot(samples_x[:200],np.zeros(samples_x.shape,dtype='float')[:200],'g.')
#####ax.plot(subx,(np.empty(subx.shape,dtype='float').fill(mean_value))[:200],color='red',linestyle='-',linewidth=3)   ??????

show()
```
