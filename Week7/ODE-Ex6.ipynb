##Thursday Example 1 : Euler Method, Free Fall

$a(t) = \frac{d^2 r}{dt^2} \quad or \quad a(t) = \frac{dv}{dt} \;, v(t) = \frac{dr}{dt}$

Given initial information $r(t_0)$ and $v(t_0)$ , I can find $v$ after one time step by using $a(t)$ and approximating it as linear, so $v(t_1) = v(t_0) + a(t_0) \Delta t$ .

General recursive formulas:

$v_{n+1} = v_n + a(t_n) \Delta t$

$r_{n+1} = r_n + \frac{1}{2} (v_{n} + v_{n+1}) \Delta t \quad$  (Using midpoint of first derivatives of $r$ on either side of the time interval)

Remember, $a(t)$ is the function we have, and $r(t)$ is what we want, but we also get $\frac{dr}{dt}$ in the process.

```python
import numpy as np

def euler_mid_odesolve(deriv,init_cond,t):
    """ deriv is a callable function that returns the value of the highest derivative of
    the unknown function at a certain t. deriv must accept inputs t, and then the first order derivative
    through the second highest derivative evaluated at t. The initial conditions go from the inital
    value of the unknown function up to the initial value of its second highest order derivative
    compared to the highest order derivative appearing in the ode, (which is deriv)
    """
    
    k = len(init_cond)
    
    result_sol = np.zeros((len(t),k + 1),dtype=float)
    
    current_vals = [0]*(k+1)
    for i in range(k):
        current_vals[i] = init_cond[i]
        result_sol[0,i] = init_cond[i]
    
    deriv_val = deriv(t[0],*current_vals[:k])
    current_vals[k] = deriv_val
    result_sol[0,k] = deriv_val
    
    
    new_vals = [0]*(k+1)
    
    for nn in range(1,len(t)):
        dt = t[nn] - t[nn-1]
        
        new_vals[k-1] = current_vals[k-1] + current_vals[k]*dt
        result_sol[nn,k-1] = new_vals[k-1]
        
        for jj in range(k-1,0,-1):
            new_vals[jj-1] = current_vals[jj-1] + (0.5)*(current_vals[jj] + new_vals[jj])*dt
            result_sol[nn,jj-1] = new_vals[jj-1]
            
        new_vals[k] = deriv(t[nn],*new_vals[:k])
        result_sol[nn,k] = new_vals[k]
        
        current_vals = new_vals
    
    
    return result_sol





def d2xdt2(t,x,dxdt):
    return -9.8
    
t = np.arange(0.,5.,0.001)
sol = euler_mid_odesolve(d2xdt2,[10.,10.],t)





fig1 = gcf()
fig1 = figure(figsize=(10.0,10.0))
ax1 = gca()

pa = ax1.plot(t,sol[:,0],'b-',label='position',lw=3.)
pb = ax1.plot(t,sol[:,1],'g-',label='vertical velocity',lw=3.)
pc = ax1.plot(t,sol[:,2],'r-',label='acceleration due to gravity',lw=3.)

title('free fall vertical of object launched upward initially',fontsize=20)
xlabel('time',fontsize=20)
ylabel('vertical distance',fontsize=20)
legend(loc=3)
grid()

show()
```
