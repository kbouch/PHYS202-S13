## #2 Weibull distribution

\begin{equation} \begin{split}
& f(x) = \begin{cases}
{\alpha \beta ^{-\alpha} x^{\alpha - 1} e^{-(x/\beta)^{\alpha}}}& x\geq 0 \\\
{0}& \text{otherwise}.
\end{cases} \\\
& \vphantom{a}\\\
& \alpha \geq 0 \;, \beta \geq 0
\end{split}
\end{equation}

__part a and b__

\begin{equation}
F(x) = \int _{- \infty } ^{x} f(t) dt = \int _{0} ^{x} \alpha \beta ^{-\alpha} t^{\alpha - 1} e^{-(t/ \beta)^{\alpha}} dt
\end{equation}

\begin{equation} \begin{split}
& \text{Let} \;\; u = \Big (\frac{t}{\beta} \Big )^{\alpha} \\\
& \vphantom{a}\\\
& du = \alpha \Big (\frac{t}{\beta} \Big )^{\alpha-1} \Big ( \frac{1}{\beta} \Big ) = \alpha \beta ^{-\alpha} t^{\alpha - 1}\\\
& \vphantom{a}\\\
& u(0) = 0\\\
& \vphantom{a}\\\
& u(x) = \Big (\frac{x}{\beta} \Big )^{\alpha}
\end{split} \end{equation}


\begin{equation}
F(x)= \int _{0} ^{u(x)} e^{-u} du = -e^{-u} \Big \vert _0 ^{u(x)} =1 -e^{- \Big ( {\displaystyle \frac{x}{\beta}} \Big ) ^{\displaystyle \alpha}}
\end{equation}

\begin{equation}
F^{-1}(z) = \beta \Big (\ln \Big ( {\displaystyle \frac{1}{1-z}} \Big ) \Big )^{\displaystyle \frac{1}{\alpha}}
\end{equation}

__part c__

```python
import numpy as np
np.random.seed(None)

a = 1.5
b = 6.

dx = 0.0001
x = np.arange(0.0,10.+ dx, dx)

def prob_distr_weibull(x,a,b):
    from scipy import integrate
    fw_notNorm = lambda x: a*(b**a)*(x**(a-1))*np.exp(-(x/b)**a)
    A = 1./integrate.trapz(fw_notNorm(x),x)
    return A*fw_notNorm(x)



def weibull_CDF_inv(z,a,b):
    return b*(np.log(1.0/(1-z)))**(1.0/a)

z = np.random.random_sample(10000)

x_samples = weibull_CDF_inv(z,a,b)

from scipy import integrate
print integrate.trapz(prob_distr_weibull(x_samples,a,b),x_samples)
```

__part d__

```python
def weibull_CDF(x,a,b):
    return 1 - np.exp(-(x/b)**a)



fig = gcf()
fig = figure(figsize(8.0,8.0))
ax = gca()

ax.hist(x_samples,100,normed=True,label = 'sampled distribution',color='blue')
ax.plot(x,prob_distr_weibull(x,a,b),color='red',lw=3.,label='true distribution')
ax.plot(x,weibull_CDF(x,a,b),color='c',lw=2,label='cumulative distribution')

ylabel(r'probability',fontsize=20)
xlabel(r'values of a quantity ',fontsize=20)

ylim(0,1.1)
xlim(0.,10.)

legend(loc = 2)
grid()
show()
```

## #3 Beta$(3,4)$ distribution

\begin{equation} \begin{split}
& f(x) = \begin{cases}
60x^{3} (1-x)^{2}& 0\leq x \leq 1 \\\
{0}& \text{otherwise}.
\end{cases} \\\
& \vphantom{a}\\\
& \alpha \geq 0 \;, \beta \geq 0
\end{split}
\end{equation}

__part a__

```python
N = 10000

def prob_distr_beta(x):
    return 60*(x**3)*(1-x)**2


x_samples_L = []
while len(x_samples_L) < N:
    x =     np.random.random_sample()
    trial = (2.0736 - 0.0)*np.random.random_sample() + (0.)
    if trial <= prob_distr_beta(x):
        x_samples_L.append(x)
    
x_samples2 = np.array(x_samples_L)
```

__part b__

\begin{equation} \begin{split}
F(x) =& \int _{-\infty} ^{x} 60 t^{3}(1-t)^{2} dt\\\
=& 60 \int _{0} ^{x} (t^{5} - 2 t^{4} + t^{3}) dt\\\
=& 10 x^{6} -24 x^{5} + 15 x^{4}\;,\; 0 \leq x \leq 1
\end{split} \end{equation}

If I sample on the interval $x \in [0,1]$ then I must normalize the original distribution with coefficient

$A = \frac{1}{F(1)} = \frac{1}{10-24+15} = \frac{1}{1} = 1$

Oh, it is already normalized.

```python
x = np.arange(0.,1.0001,0.0001)

def beta_CDF(x):
    return 10*(x**6) - 24*(x**5) + 15*(x**4)


fig = gcf()
fig = figure(figsize(8.0,8.0))
ax = gca()

ax.hist(x_samples2,100,normed=True,label = 'sampled distribution',color='blue')
ax.plot(x,prob_distr_beta(x),color='red',lw=3.,label='true distribution')
ax.plot(x,beta_CDF(x),color='c',lw=2,label='cumulative distribution')

ylabel(r'probability',fontsize=20)
xlabel(r'values of a quantity ',fontsize=20)

ylim(0,3.0)
xlim(-.2,1.2)

legend(loc = 2)
grid()
show()
```
