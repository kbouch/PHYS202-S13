###Cumulative Distribution Function (CDF)

####Example 1

```python
x = numpy.arange(0.,numpy.pi + 0.01,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (1-numpy.cos(x))/2.
    
plot(x,f(x),label='True distribution',linewidth=3.)
plot(x,F(x),label='Cumulative function',linewidth=3.)

xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
N = 10000
```

```python
z = numpy.random.random_sample(N)

Finv = lambda z: numpy.arccos(1-(2*z))

xsamples = Finv(z)

hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f(x),label='True distribution',lw=3.,color='red')

xlim(0,numpy.pi)
ylim(0,0.7)

legend(loc=0)
show()
```

####Exercise 1 : exponential function - analytic

\begin{equation}
f(x) = \lambda e^{ -\lambda x}
\end{equation}

$x \in [0,5]$

cumulative distribution function: $z = F(x) = \int _0 ^x \lambda e^{ -\lambda x} dx \quad = e^{-\lambda x} - 1$

and its inverse: $x = F^{-1}(z) =\frac{-1}{\lambda} \ln (z + e^{- \lambda*(0)}) = \frac{-1}{\lambda} \ln (1-z)$

now let $\lambda = 1$ and plot the actual distribution versus sampled distribution.


```python
x = numpy.arange(0.0,5.01,0.01)

f = lambda x: numpy.exp(-x)
    
N = 10000

F = lambda x: numpy.exp(-x) -1
    
Finv = lambda z: -1.*numpy.log(1-z)
    
z = numpy.random.random_sample(N)

xsamples = Finv(z)

binned = hist(xsamples,50,normed=True,label = 'sampled distribution',color='blue')
plot(x,f(x),color='red',lw=3.,label='true distribution')

xlim(0.,5.5)
ylim(0.,1.1)

legend(loc=0)

show()
```
