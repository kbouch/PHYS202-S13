####Exercise 1 : exponential function - analytic

\begin{equation}
f(x) = \lambda e^{ -\lambda x}
\end{equation}

$x \in [0,5]$

cumulative distribution function: $z = F(x) = \int _0 ^x \lambda e^{ -\lambda x} dx \quad = e^{-\lambda x} - 1$

and its inverse: $x = F^{-1}(z) =\frac{-1}{\lambda} \ln (e^{- \lambda*(0)} - z) = \frac{-1}{\lambda} \ln (1-z)$

now let $\lambda = 1$ and plot the actual distribution versus sampled distribution.


```python
x = numpy.arange(0.0,5.01,0.01)

A = 1./(1-np.exp(-5))

f = lambda x: A*numpy.exp(-x)
    
N = 10000

F = lambda x: numpy.exp(-x) -1
    
Finv = lambda z: -1.*numpy.log(1-z)
    
z = numpy.random.random(N)

xsamples = Finv(z)

hist(xsamples,50,normed=True,label = 'sampled distribution',color='blue')
plot(x,f(x),color='red',lw=3.,label='true distribution')

xlim(0.,5.5)
ylim(0.,1.1)

legend(loc=0)

show()
```

```python
import numpy as np
from scipy import integrate
```

```python
dx = 0.001
x = numpy.arange(0.0,5. + dx, dx)

f_notNorm = lambda x: numpy.exp(-x)

#A = 1./(1-np.exp(-5.))
A = 1./integrate.trapz(f_notNorm(x),x)

f = lambda x: A*numpy.exp(-x)
    
f_eval = f(x)

F = np.ones(x.shape,dtype=float)

for i in range(x.shape[0]):
    F[i] = integrate.trapz(f_eval[:i+1],x[:i+1])


N = 10000
np.random.seed(None)
z = np.random.random_sample(N)


def Finv(z,F,x):
    """ Takes an array z of randomly sampled floats on the interval 
    [0.,1.) such as with numpy.random.random(10000). Also takes an array of 
    values of the CDF transform function F evaluated at each value in
    an array x. x is also passed. F and x have the same shape.
    Returns the x values by taking the inverse transform of the z
    (random floats from 'continous uniform' distribution which are spread 
    over F (cumulative probability density of f)). Returns x_samples
    that have probability given by f, upon which F is based.
    """
    x_indicies = np.searchsorted(F,z)
    x_samples = []
    for ind in x_indicies:
        x_samples.append(x[ind])
    return np.array(x_samples,dtype=float)

x_samples = Finv(z,F,x)


binned = hist(x_samples,50,normed=True,label = 'sampled distribution',color='blue')
plot(x,f(x),color='red',lw=3.,label='true distribution')
plot(x,F,color='c',lw=2,label='cumulative distribution')

xlim(0.,5.)
ylim(0.,1.1)

legend(loc=0)

show()
```

```python
def Finv(z,F,x):
    """ Takes an array z of randomly sampled floats on the interval 
    [0.,1.) such as with numpy.random.random(10000). Also takes an array of 
    values of the CDF transform function F evaluated at each value in
    an array x. x is also passed. F and x have the same shape.
    Returns the x values by taking the inverse transform of the z
    (random floats from 'continous uniform' distribution which are spread 
    over F (cumulative probability density of f)). Returns x_samples
    that have probability given by f, upon which F is based.
    """
    x_indicies = np.searchsorted(F,z)
    x_samples = []
    for ind in x_indicies:
        x_samples.append(x[ind])
    return np.array(x_samples,dtype=float)


def distribute1D(x,probx,N,zseed=None):
    """Takes an array x of values of a quantity and a second array of the same length
    giving the relative probability of corresponding values in x to occur. Generates
    N random vaules with same distribution given by probx, with outputs landing only on
    values given in x (so make the bins narrow enough)."""
    
    from scipy import integrate
    import numpy.random as rand
    
    A = 1.0/integrate.trapz(probx,x)
    
    F = np.ones(x.shape[-1],dtype=float)
    
    for i in range(x.shape[-1]):
        F[i] = A*integrate.trapz(probx[:i+1],x[:i+1])
    
    rand.seed(zseed)
    z = rand.random_sample(N)
    
    return Finv(z,F,x)
```

```python
import numpy as np
xstep = 0.0001
x = np.arange(-0.1,0.1+xstep,xstep) #meters from screen center

def intensity(x,a,d,lmbda,L):
    import numpy as np
    ISS = (np.sin(a*x/(lmbda*L))/(np.pi*x))**2.0
    IDS = (np.cos(2*np.pi*d*x/(lmbda*L)))**2.0
    return ISS*IDS

A = 1.0/integrate.trapz(intensity(x,a,d,lmbda,L),x)

d = 15e-6 #meters
a = 10e-6 #meters
lmbda = 632.8e-9 #meters
L = 1  #meters

photon_strikes_x = distribute1D(x,intensity(x,a,d,lmbda,L),10000)

hist(photon_strikes_x,100,normed=True,label = 'sampled distribution',color='blue')
plot(x,A*intensity(x,a,d,lmbda,L),color='red',lw=3.,label='true distribution')

xlabel('distance from screen center (m)',fontsize=20)
ylabel('intensity',fontsize=20)

xmin,xmax = xlim(-0.1,0.1)

show()
```
