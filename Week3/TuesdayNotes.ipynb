Create a very simple plot

```python
import numpy as np
t = np.arange(0.0,1.0+0.01,0.01)
pi = np.pi
c = np.cos(4*pi*t)
s = np.sin(4*pi*t)
# np.cos accepts an array, ouputs an array of the same shape 
# with each elemement of argument array evaluated in cosine function
```

```python
plot(t, s, color='blue',linewidth=2.5,linestyle='-',label='$cos(4{/pi}t)$')
plot(t, c, color='red',linewidth=2.5,linestyle='--',label='$sin(4{/pi}t)$')
xlabel('time (s)')
ylabel('voltage (mV)')
# Add title for convenience
title('A simple plot',fontsize=20)
grid(True)
legend(loc='lower left')
savefig('simple_plot.pdf')
```
