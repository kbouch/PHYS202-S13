##Line plots

```python
import numpy as np
D = np.arange(-5,5.01,0.01) # Array containing many values spread throughout a domain interval with small increment between subsequent values.
                            # evalute with a numpy math function to get a smooth graph if the resulting array and domain array are plotted together.

E = np.cosh(D)  # The numpy version of hyperbolic cosine evalutes the function for each array entry 
                # and stores the ouputs in an array of the same shape.

p = np.sinh(D)  # Also store the results of evaluating hyperbolic sine over the domain.
```

```python
# pylab is inline (or is it matplotlib.pyplot??????)
plot(D, E, 'b-', color='b',             linewidth=3.0, linestyle='-', label='$E$') # color and linestyle can be given as a third *arg instead
                                                                                      # of using keywords. In this way you can plot multiple
                                                                                      # x-y pairs of arrays as a set of 3 *args, but the keywords
                                                                                      # apply to all data sets, limiting customization.
plot(D, p,       color=(0,0.784,0.216), linewidth=3.0, linestyle='-', label='$|p|$')

title('E and $|p|$ versus D',fontsize=15)      # optional, block with comment symbol
legend(loc='upper right')         
myxlabel = xlabel('$x$ (rad)', fontsize=15)    # x axis label, assigned and stored
myylabel = ylabel('out values', fontsize=15)   # y axis label

ymax = np.max(np.array([np.max(E),np.max(p)]))   # find the maximum y value attained by any plotted line
ymin = np.min(np.array([np.min(E),np.min(p)]))   # find the minimus y value
                                                 # this will be used for tick spacing and labeling

xmin, xmax = xlim(-5,5)      # pyplot.xlim sets the axis limits, but also returns the current limits.
ymin, ymax = ylim(ymin,ymax) # I will leave this auto scaled, unless I need it.
                             # May need to adjust axis bounds in order to fit legend

xtickstep = 1.0
myxlocs = np.arange(xmin,xmax+xtickstep,xtickstep,dtype='float')  # using variables assigned previously initiate arguments for the pyplot.xticks function
myxticklabels = np.array(myxlocs[:], dtype='string')

ytickstep = (ymax - ymin)/10.0
myylocs = np.arange(ymin,ymax+ytickstep,ytickstep,dtype='float')
myyticklabels = np.array(myylocs[:],dtype='string')

(a,b) = plt.xticks(myxlocs,myxticklabels)                            # plug in variables created above, assign the return value to something to
(c,d) = plt.yticks(myylocs,myyticklabels)                            # avoid notebook printing object references.


ax1 = gca()
ax1.spines['right'].set_color('None')
```
