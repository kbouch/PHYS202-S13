##Line plots

```python
import numpy as np
D = np.arange(-5,5.01,0.01) # Array containing many values spread throughout a domain interval with small increment between subsequent values.
                            # evalute with a numpy math function to get a smooth graph if the resulting array and domain array are plotted together.

E = np.cosh(D)  # The numpy version of hyperbolic cosine evalutes the function for each array entry 
                # and stores the ouputs in an array of the same shape.

p = np.sinh(D)  # Also store the results of evaluating hyperbolic sine over the domain.
```

```python
import matplotlib.pyplot as plt
# return an instance of :class:matplotlib.figure.Figure
fig = plt.figure(num=1,figsize=(10.0,10.0),frameon=False,dpi=80)
ax = fig.add_axes([0,1,1,1],frameon=False)


ax.plot(D, E, 'b-', color='blue',          linewidth=2.0, linestyle='-', label='$E$') # color and linestyle can be given as a third *arg instead
                                                                                      # of using keywords. In this way you can plot multiple
                                                                                      # x-y pairs of arrays as a set of 3 *args, but the keywords
                                                                                      # apply to all data sets, limiting customization.
ax.plot(D, p,       color=(0,0.784,0.216), linewidth=2.0, linestyle='-', label='$|p|$')

plt.title('E and $|p|$ versus D',fontsize=15)      # optional, block with comment symbol
plt.legend(loc='upper right')         
my_xlabel = plt.xlabel('$x$ (rad)', fontsize=15)    # x axis label, assigned and stored
my_ylabel = plt.ylabel('out values', fontsize=15)   # y axis label

print my_xlabel
print my_ylabel

ymax = np.max(np.array([np.max(E),np.max(p)]))   # find the maximum y value attained by any plotted line
ymin = np.min(np.array([np.min(E),np.min(p)]))   # find the minimus y value
                                                 # this will be used for tick spacing and labeling

xmin, xmax = plt.xlim(-5,5)      # pyplot.xlim sets the axis limits, but also returns the current limits.
ymin, ymax = plt.ylim(ymin,ymax) # I will leave this auto scaled, unless I need it.
                                 # May need to adjust axis bounds in order to fit legend

plt.grid(b=True, which='major', axis='both')
    
xtickstep = 1.0
my_xlocs = np.arange(xmin,xmax+xtickstep,xtickstep,dtype=np.float32)  # using variables assigned previously initiate arguments for the pyplot.xticks function
my_xticklab = np.array(my_xlocs[:], dtype=np.str_)

ytickstep = (ymax - ymin)/10.0
my_ylocs = np.arange(ymin,ymax+ytickstep,ytickstep,dtype=np.float32)
my_yticklab = np.array(my_ylocs[:],dtype=np.str_)

(a,b) = plt.xticks(my_xlocs,my_xticklab)                            # plug in variables created above, assign the return value to something to
(c,d) = plt.yticks(my_ylocs,my_yticklab)                            # avoid notebook printing object references.

#print ax.spines # axes class attribute is a dictionary with values that are :class:matplotlib.spines.Spine
#print ax.spines.keys()
#print ax.spines.values()

# access each Spine object located in the spines dictionary
ax.spines['top'].set_color('none')            # set the top and bottom spine color to none to make them invisible
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position( ('data',0.0) ) # takes one tuple as an arguemnt
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position( ('data',0.0) )

plt.show()
```
